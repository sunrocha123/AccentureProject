## Comparação do número de casos do Brasil com outros três 'epicentros' da COVID-19

```
import pyspark.sql.functions as f

df = df_data_by_country.select('NOME', 'CONFIRMED', 'DATE') \
      .where('NOME in("Brazil", "China", "Italy", "United States of America")') \
      .orderBy(f.asc('DATE'))

display(df)
```
